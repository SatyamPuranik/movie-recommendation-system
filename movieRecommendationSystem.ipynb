```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
movies.columns.values
movies = movies[['id','title','genres','keywords','overview','tagline']]
movies.head()
movies.isna().sum()
```

```python
credits.head()
credits.rename(columns={'movie_id':'id'},inplace=True)
credits.drop(columns='title',inplace =True)
credits.head()
```

```python
movies.shape
```

```python
credits.shape
```

```python
df = movies.merge(credits,on='id')
df.head()
```

```python
df.loc[df['tagline'].isna(),'tagline'] = ''
df.loc[df['overview'].isna(),'overview'] = ''
df.isna().sum()
```

```python
import ast
def convert(obj) -> list:
    L =[]
    for i in ast.literal_eval(obj):
         L.append(i['name'])
    return ' '.join(L)
```

```python
df.genres = df.genres.apply(convert)
df.keywords = df.keywords.apply(convert)
df
```

```python
def director(obj):
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            return i['name']
    return ''

def actors(obj):
    L =[]
    for i in ast.literal_eval(obj):
        if len(L) >=3:
            break
        else:
         L.append(i['name'])
    return ' '.join(L)
```

```python
df.cast = df['cast'].apply(actors)
df.crew = df.crew.apply(director)
df
```

```python
df = df[['id','title','overview','tagline','genres','keywords','cast','crew']]
df
```

```python
df['Vector'] = df['overview']+df['tagline']+df['genres']+df['keywords']+df['cast'] +' '+ df['crew']
df.head()
```

```python
df = df[['id','title','Vector']]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text:str)->str:
    x =[]
    for i in text.split():
        x.append(ps.stem(i))
    return ' '.join(x)
```

```python
df.Vector = df.Vector.apply(stem)
df.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(df.Vector)
vector = vector.toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
similarity
```

```python
def recommendation(movie:str):
    index = df[df.title == movie].index[0]
    dist = similarity[index]
    l = sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])
    for i in l[1:6]:
        # print(df.iloc[i[0]].title)
        print(i)
```

```python
recommendation('Batman')
```
